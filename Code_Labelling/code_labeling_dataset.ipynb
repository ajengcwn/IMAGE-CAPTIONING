```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import os
```

```python
# Cek apakah file benar-benar ada di lokasi tersebut
file_path = '../data/raw/dataset_clean.csv'
if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(f"Berhasil memuat {len(df)} data dari: {file_path}")
else:
    print(f"Gagal! File tidak ditemukan di: {os.path.abspath(file_path)}")
    print("Pastikan folder 'data/raw' berada satu level di luar folder 'Code_Labelling'")
```

```python
# 2. Inisialiasai Model AI

print("Memuat model dan memproses embedding...")
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# 3. Proses Embedding (Mengubah teks menjadi angka)

embeddings = model.encode(df['caption_desc'].tolist(), show_progress_bar=True)
```

```python
# 4. Clustering dengan K-Means
num_clusters = 4
kmeans = KMeans(n_clusters=num_clusters, n_init='auto', random_state=42)
clusters = kmeans.fit_predict(embeddings)
```

```python
# 5. Mapping Mood Sesuai Permintaan
mood_map = {
    0: "Joy",
    1: "Sad",
    2: "Surprise",
    3: "Love"
}

# Update kolom emotion_label
df['emotion_label'] = [mood_map[c] for c in clusters]
```

```python
# 6. Simpan Hasil
from pathlib import Path

# Definisikan ulang path-nya
current_dir = Path().absolute()

# 6. Simpan Hasil
output_path = current_dir.parent / 'data' / 'raw' / 'dataset_with_mood.csv'
df.to_csv(output_path, index=False)

print(f"Proses selesai! Hasil disimpan di: {output_path}")
```
